```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
df = pd.read_csv('/content/drive/MyDrive/Projects/Project 2 : Diabetes prediction/diabetes.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df['Outcome'].value_counts()
```

```python
df.groupby('Outcome').mean()
```

```python
X = df.drop(columns='Outcome',axis=1)
Y = df['Outcome']
```

```python
print(Y)
```

```python
scaler = StandardScaler()
standard = scaler.fit(X)
standardized_data = standard.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , stratify = Y , random_state=2 ,test_size =0.2)
```

```python
print(X.shape , X_train.shape , X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train,Y_train)
```

```python
X_train_prediction = classifier.predict(X_train)
training_prediction = accuracy_score(X_train_prediction,Y_train)
```

```python
print(training_prediction)
```

```python
X_test_prediction = classifier.predict(X_test)
testing_prediction = accuracy_score(X_test_prediction,Y_test)
```

```python
print(testing_prediction)
```

```python
input_data = (2,197,70,45,543,30.5,0.158,53)

data =  np.asarray(input_data)
reshaped_data = data.reshape(1,-1)

std_data = scaler.transform(reshaped_data)
prediction = classifier.predict(std_data)
print(prediction)

if(prediction[0]==0):
  print("Not Diabetic")
else:
  print("Diabetic")
```

```python
df.replace({'BusinessTravel':{'Travel_Rarely':1,'Non-Travel':0,'Travel_Frequently':2},'Department':{'Research & Development':0,'Sales':1,'Human Resources':2},
            'Gender':{'Male':0,'Female':1},'Status_of_leaving':{'Dept.Head':0,'Salary':1,'Work Environment':2,'Work Accident':3,'Better Opportunity':4},
            'Mode_of_work':{'WFH':0,'OFFICE':1},'Work_accident':{'No':0,'Yes':1},'Source_of_Hire':{'Recruiter':0,'Job Event':1,'Walk-in':2,'Job Portal':3},
            'Job_mode':{'FullTime':0,'Contract':1,'Part Time':2},'OverTime':{'No':0,'Yes':1},'MaritalStatus':{'Married':0,'Single':1,'Divorced':2},
            'Higher_Education':{'Post-Graduation':0,'Graduation':1,'PHD':2,'12th':3},'Attrition':{'Yes':1,'No':0}},inplace=True)
```

```python
df = df.drop(columns='JobRole',axis=1)
df = df.drop(columns='Date_of_Hire',axis=1)
```

```python
df['OverTime']
```

```python
import numpy as np

df.groupby('Attrition').mean()
```

```python
X = df.drop(columns='Attrition',axis=1)
Y = df['Attrition']
```

```python
print(Y)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training the model:

Support Vector Machine Model

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support Vector Macine model
classifier.fit(X_train,Y_train)
```

Model Evaluation

```python
# accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuray = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data : ', training_data_accuray)
```

```python
# accuracy score on training data
X_test_prediction = classifier.predict(X_test)
test_data_accuray = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data : ', test_data_accuray)
```
